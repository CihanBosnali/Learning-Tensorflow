## Lecture 2 - Graphs in TF ##



```python
import tensorflow as tf
```

### Default Graph ###
When you create a tf object, it initializes a default graph.

```python
graph = tf.get_default_graph()
```

### Display Operations using Graph###

```python
graph.get_operations()
```

### Build a Simple Graph ###
We will just write some operations...

```python
a = tf.constant(7, name="a")
b = tf.constant(4, name="b")
c = tf.add(a,b)
d = tf.multiply(a, c)
```

### Display Operations using Graph###
Now, we will have something in our graph

```python
ops = graph.get_operations()
print(ops)
```

### Let's show more info about operations ###

```python
for operation in ops:
    print("------ Start of Operation ------\n")
    print(operation)
    print("------ End of Operation ------\n")
```

### Run the TF Program ###

```python
with tf.Session() as session:
    result = session.run(d)
    print(result)
```
